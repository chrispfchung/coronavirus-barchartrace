```python
# https://towardsdatascience.com/bar-chart-race-in-python-with-matplotlib-8e687a5c8a41
# # bad dataset - https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.animation as animation
from IPython.display import HTML
import tqdm
%matplotlib inline
```

```python
# load in dataset
# this is a good dataset that broke down the time stamps into M/D/YY
# It also has 5678 data points for all countries affected by COVID-19
df = pd.read_excel('COVID-19-geographic-disbtribution-worldwide-2020-03-19.xlsx')
print(df.info()) 
df.head(3)
```

```python
# I like to lower case and add underscores to my column names for easier access
df.columns = [x.lower() for x in df.columns]
df.columns = [x.replace(' ', '_') for x in df.columns]
df.columns
```

```python
# okay so I'm sorting my dataframe by month, day then cases, in ascending order.
# then I'm reversing the values to get the most cases at the top and doing a double reverse so that our cont.
# top number appears at the top of our horizontal bar graph. Shown in following code.

# set starting date equal to
beginning_date = df['daterep'].sort_values(ascending=True).head(1).item()

#create starting bar chart race from the first day
dff = (df[df['daterep'].eq(beginning_date)]
       .sort_values(by='cases', ascending=False)
       .head(10))

dff
```

```python
# plot out horizontal bar chart
fig, ax = plt.subplots(figsize=(15,8))
ax.barh(dff['countries_and_territories'], dff['cases'])
```

```python
colors =  ['#adb0ff', '#ffb3ff', '#90d595', '#e48381',
     '#aafbff', '#f7bb5f', '#eafb50', '#adb0ff', '#ffb3ff']

group_lk = dff.set_index('countries_and_territories')['cases'].to_dict()
group_lk
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
# dff = dff[::-1]   # flip values from top to bottom
# pass colors values to `color=`
ax.barh(dff['countries_and_territories'], dff['cases'], color=colors)

# iterate over the values to plot labels and values (Tokyo, Asia, 38194.2)

for i, (name, value) in enumerate(zip(dff['countries_and_territories'], dff['cases'])):
#     ax.text(value, i,     name,            ha='right')  # Tokyo: name
    ax.text(value, i, group_lk[name],  ha='right')  # Asia: group name
#     ax.text(value, i,     value,           ha='left')   # 38194.2: value
# Add year right middle portion of canvas
ax.text(1, 0.4, 2020, transform=ax.transAxes, size=46, ha='right')
```

```python
# datetime(2019,12,31) - datetime(2019,12,31) This works!
days_since = abs((df['daterep'].sort_values(ascending=True).head(1).item() - 
                  df['daterep'].sort_values(ascending=False).head(1).item()).days)
days_since
```

```python
from datetime import date
from datetime import datetime
beginning_date = df['daterep'].sort_values(ascending=True).head(1).item()
date_list = pd.date_range(beginning_date, periods=days_since).to_list() # create a range of date
type(date_list)
```

```python
def draw_barchart(daterep):
    fig, ax = plt.subplots(figsize=(15, 8))

    dff = df[df['daterep'].eq(daterep)].sort_values(by='cases', ascending=False).head(15).reset_index()
    ax.clear() # you need this to help you refresh the axes for the next graphic

    ax.barh(dff['countries_and_territories'], dff['cases'], color=colors)
    for i, (value, name) in enumerate(zip(dff['cases'], dff['countries_and_territories'])):
        print(i, (value,name))
#         ax.text(value, i,     name,           size=10, weight=600, ha='right', va='center') # country on bar
        ax.text(value, i,     f'{value:,.0f}',  size=12, weight=600,ha='left',  va='center') #number on bar
#     # ... code to polish style below ... #
    ax.text(1, 0.4, daterep.date(), transform=ax.transAxes, 
            color='#777777', size=46, ha='right', 
            weight=800) #toggle big current date on white space

    ax.text(0, 1.06, 'Cases', transform=ax.transAxes, size=12, color='#000000') #sets cases word in top corner
#     ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}')) 
    ax.xaxis.set_ticks_position('top') # sets tick positions on top of plot
    ax.tick_params(axis='x', colors='#777777', labelsize=20) # adjust size of x labels (ticks)
    ax.tick_params(axis='y', colors='#000000', labelsize=15) # adjust size of y labels (ticks)
#     ax.set_yticks([]) # removes y labels
    ax.margins(0, 0.01) #sets space between bars
    ax.grid(which='major', axis='x', linestyle='-') # sets vertical lines as place markers through race
    ax.set_axisbelow(True) 

    
    ax.text(0, 1.12, 'Reported Cases of Coronavirus by Country Since New Year\'s Eve',
            transform=ax.transAxes, size=24, weight=600, ha='left')

    
# #     ax.text(1, 0, 'by @pratapvardhan; credit @jburnmurdoch', transform=ax.transAxes, ha='right',
#             color='#777777', bbox=dict(facecolor='white', alpha=0.8, edgecolor='white'))
    plt.box(False)
    
    
draw_barchart(datetime(2019,12,31)) # ascending = false. and head -- Correct!
```

```python
draw_barchart(datetime(2019,12,31)) # ascending = false. and head -- Correct!
```

```python
draw_barchart(datetime(2020,1,4))
```

```python
import matplotlib.animation as animation
from IPython.display import HTML
fig, ax = plt.subplots(figsize=(15, 8))
animator = animation.FuncAnimation(fig, draw_barchart, 
                                   frames=date_list[5:15])

HTML(animator.to_jshtml())  # animator tool bar
```
